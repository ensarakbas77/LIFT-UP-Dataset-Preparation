```python
from pathlib import Path

# Excel dosyalarının olduğu klasör
folder_path = Path(".")

# Klasördeki tüm xlsx dosyalarını al
files = sorted(folder_path.glob("20*-20*.xlsx"))

# Okuyup birleştir
dfs = [pd.read_excel(file) for file in files]

merged_df = pd.concat(dfs, ignore_index=True)
```

```python
merged_df.shape
```

```python
merged_df.isnull().sum()
```

```python
year_counts = merged_df["Year"].value_counts().sort_index()

for year, count in year_counts.items():
    print(f"{year} Yılı Toplam Kayıt Sayısı: {count}")
    print("-" * 40)
```

```python
# Kaydetme
merged_df.to_excel("all_articles.xlsx", index=False)
merged_df.to_csv("all_articles.csv", index=False, encoding="utf-8")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Stil
sns.set_theme(style="whitegrid", font_scale=1.1)

plt.figure(figsize=(6, 4))

ax = sns.countplot(
    data=articles,
    x="Year"
)

# Bar üstüne sayıları yaz
for p in ax.patches:
    ax.annotate(
        f"{int(p.get_height())}",
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha="center",
        va="bottom",
        fontsize=11,
        xytext=(0, 4),
        textcoords="offset points"
    )

ax.set_title("Yıllara Göre Makale Sayısı", fontsize=14, pad=12)
ax.set_xlabel("Yıl")
ax.set_ylabel("Makale Sayısı")

sns.despine()
plt.tight_layout()
plt.show()
```
